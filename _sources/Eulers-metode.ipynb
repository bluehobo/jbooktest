# Eulers metode - fritt fall

Simulerer fritt fall, m=75kg. k er funnet eksperimentelt for terminalhastighet p√• ca 200 km/h

```python
# -*- coding: utf-8 -*-
"""
Created on Wed Mar  8 19:11:34 2023

Simulasjon av fritt fall for fallskjermhopper

@author: Sigurd Rage - USN
"""

# Importerer biblioteker
import numpy as np
import matplotlib.pyplot as plt

# Initierer step dt og totaltid T for simulasjonen
dt = 0.1
T=25

# Initierer arrays ut fra betingelsene over
N = int(T/dt)+1

t = np.zeros(N)
s = np.zeros(N)
v = np.zeros(N)

# Initierer starth√∏yde s og starthastighet v
s[0] = 0.0
v[0] = 0.0

# Definerer masse, tyngdeakselerasjon og drag-koeffisenten k
m = 75
g = 9.81
k = 0.23835

# Definerer funksjon for beregning av total akselerasjon
def dv_dt(fart):
    return g - (k * fart**2 / m)
        
# Kj√∏rer simulasjonen
for i in range(N-1):
    # beregner n√•v√¶rende akselerasjon fra funksjonen dv_dt
    a = dv_dt(v[i])
    # beregner fart og posisjon for neste element i simulasjoner
    v[i+1] = v[i] + a*dt
    s[i+1] = s[i] + v[i]*dt + 0.5*a*dt**2
    # beregner t verdi for neste element
    t[i+1] = t[i] + dt

# plotter grafer for posisjon og tid i simulasjonen
plt.close("all")

plt.figure()
plt.plot(t,s)
plt.xlabel("Tid (s)")
plt.ylabel("Posisjon (m)")
plt.show()

plt.figure()
plt.plot(t,v)
plt.xlabel("Tid (s)")
plt.ylabel("Fart (m/s)")
plt.show()

# skriver ut hastighet for siste element i simulasjonen, for statisk test
print(str(v[v.size-1]*3.6) + " km/h")
```
